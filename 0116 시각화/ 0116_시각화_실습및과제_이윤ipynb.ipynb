```python
# 여기에 시각화 코드 혹은 태블로 public 링크를 첨부해주세요.
df = pd.read_csv('/content/StockX-Data-Contest-2019-3.csv')
df.head()
```

### EDA part

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 데이터 타입 확인
df.dtypes
```

```python
# 데이터 타입 확인 결과 가격에 해당되는 'sale price'와 'retail price'의 데이터 형태가 'object'인 것을 확인할 수 있음
# "$", ","를 없애고 'sale price'와 'retail price'의 데이터타입을 int로 변경
df['Sale Price'] = pd.to_numeric(df['Sale Price'].str.replace('$', '').str.replace(',', ''), errors='coerce').astype(int)
df['Retail Price'] = pd.to_numeric(df['Retail Price'].str.replace('$', '').str.replace(',', ''), errors='coerce').astype(int)
```

```python
# 데이터 타입이 변경된 것을 확인할 수 있다
print(df.dtypes)
print(df.head)
```

```python
# 브랜드에 따른 판매량을 확인
brand_sales = df.groupby('Brand').size().reset_index(name='Total Sales')
sns.barplot(x='Brand', y='Total Sales', data=brand_sales)
plt.xlabel('Brand')
plt.ylabel('Total Sales')
plt.title('Total Sales Count by Brand')
plt.xticks(rotation=45)
plt.show()
```

```python
# 신발 종류에 따른 판매량을 확인
sneaker_sales = df.groupby('Sneaker Name').size().reset_index(name='Total Sales')
sneaker_sales = sneaker_sales.sort_values(by='Total Sales', ascending=False)
plt.figure(figsize=(10, 6))
palette = sns.color_palette("viridis", len(sneaker_sales))
plt.bar(sneaker_sales['Sneaker Name'], sneaker_sales['Total Sales'], color=palette)

plt.xlabel('Sneaker Name')
plt.ylabel('Total Sales')
plt.title('The number of Traded Sneakers')
plt.xticks(fontsize=9)
plt.xticks(rotation=90)

plt.show()
```

```python
# 판매가격과 소매가격의 분포에 대한 확인
# 판매가격에 대한 분포
plt.figure(figsize=(8, 3))
plt.hist(df['Sale Price'], bins=20, color='pink', alpha=0.7, label='Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.title('Histogram of Sale Price')
plt.legend()

# 소매가격에 대한 분포
plt.figure(figsize=(8, 3))
plt.hist(df['Retail Price'], bins=5, color='green', alpha=0.7, label='Retail Price')
plt.xlabel('Retail Price')
plt.ylabel('Frequency')
plt.title('Histogram of Retail Price')
plt.legend()

plt.show()
```

### 시각화 part

```python
f, ax = plt.subplots(figsize=(12, 6.5))
sns.despine(f, left=True, bottom=True)
# 프리미엄 가격에 대한 열을 생성
df['premium'] = df['Sale Price']-df['Retail Price']
# 판매량이 많은순으로 리스트 생성
brand_counts = df['Brand'].value_counts().index.tolist()
# 신발 사이즈에 따른 프리미엄의 관계를 알수있도록 그래프를 그림
# 위와 더불어 가격에 영향을 미칠 수 있는 브랜드를 색상을 통해 알수있도록 함
sns.scatterplot(x="Shoe Size",
                y="premium",
                hue="Brand",
                palette="viridis",
                sizes=(10, 200),
                alpha=0.7,
                linewidth=0.5,
                data=df,
                size = 4,
                ax=ax)
legend = plt.legend()
legend.get_texts()[2].set_text('')
plt.title('Price premium based on the size of shoes')
```
